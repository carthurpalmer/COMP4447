```python
import pandas as pd
import numpy as np
```

```python
zillow_df = pd.read_csv('zillow_df.csv')
zillow_data = pd.read_csv('zillowdatacsv.csv')
```

```python
zillow_df.head()    
```

```python
zillow_data.head()
```

```python
roi_df = pd.merge(zillow_df, zillow_data, on='zpid', how='inner')
```

```python
roi_df = roi_df[['price_x', 'propertyTaxRate_x', 'streetAddress_x', 'city', 'state', 'country', 'pa_rev', 'pa_cost', 'rf_hoaFee']]
```

```python
roi_df
```

```python
roi_df['down_pmt_pct'] = 0.05
```

```python
roi_df['down_pmt'] = roi_df['price_x'] * roi_df['down_pmt_pct']
```

```python
roi_df.head()
```

```python
roi_df['hoa_split'] = roi_df['rf_hoaFee'].apply(
    lambda x: x.split(' ') if isinstance(x, str) else x
)
```

```python
roi_df['hoa_amt'] = roi_df['hoa_split'].apply(
    lambda x: int(x[0].replace('$', '').replace(',', '')) if isinstance(x, list) else int(x.replace('$', '').replace(',', '')) if isinstance(x, str) else x
)
```

```python
roi_df['hoa_freq'] = roi_df['hoa_split'].apply(
    lambda x: x[1] if isinstance(x, list) else x
)
```

```python
roi_df['pa_hoa'] = roi_df.apply(lambda row: row['hoa_amt'] * 12 if row['hoa_freq'] == 'monthly' else row['hoa_amt'], axis=1)        
```

```python
roi_df['pa_mortgage'] = 12 * ((roi_df['price_x'] - roi_df['down_pmt']) * (roi_df['interest_rate'] / 12)) / (1 - ((1 + (roi_df['interest_rate'] / 12)) ** (-360)))
```

```python
roi_df['y1_cf'] = -roi_df['down_pmt'] - roi_df['pa_mortgage'] - roi_df['pa_cost']  - roi_df['pa_hoa'] + roi_df['pa_rev']
roi_df['y2-y15_cf'] = -roi_df['pa_mortgage'] - roi_df['pa_cost']  - roi_df['pa_hoa'] + roi_df['pa_rev']
```

```python
import numpy_financial as npf
```

```python
roi_df['cash_flows'] = roi_df.apply(lambda row: [row['y1_cf']] + [row['y2-y15_cf']] * 14, axis=1)
```

```python
roi_df['irr'] = roi_df.apply(lambda row: npf.irr(row['cash_flows']), axis=1)
```
